```python
f=open("9.txt")
k=0
for i in f:
    n=list(map(int,i.split()))
    n.sort()
    if (n[0]+n[4])**2>n[1]**2+n[2]**2+n[3]**2:
        k+=1
print(k)
```

```python
f=open("9_2.txt").readlines()
k=0
x=[]
for i in f:
    n=list(map(int, i.split()))
    x.extend(n)
for p in f:
    n=list(map(int, p.split()))
    for el in n:
        if n.count(el)==1 and x.count(el)==46:
            k+=1
print(k)

```
